```python
# Seleccionar centros que sexan de titularidade privada. Cantos hai?

resultado1_df = df[df['TITULARIDADE'] == 'Privada']

print(f"Hay {len(resultado1_df.index)} centros de titularidade privada.")
resultado1_df
```

```python
# Seleccionar centros que teñan ensinanzas concertados. Cantos hai?

resultado2_df = df[df['ENSINO_CONCERTADO'] == 'Si']

print(f"Hay {len(resultado2_df.index)} centros concertados.")
resultado2_df
```

```python
# Hai algún centro con ensinanzas concertadas dependente da Consellería?

resultado3_df = df[(df['ENSINO_CONCERTADO'] == "Si") & (df['DEPENDENTE'] == "Si")]

resultado3_df

if(len(resultado3_df.index) == 0):
    print("Non hai centros con ensinanzas concertadas dependentes da Cosellería.")
else:
    print (f"Hai {len(resultado3_df.index)} centros con ensinanzas concertadas dependentes da Cosellería.")
```

```python
# Seleccionar centros de titularidade pública de Coruña. Cantos hai?

resultado4_df = df[(df['TITULARIDADE'] == "Pública") & (df['Provincia'] == "A Coruña")]


print(f"Hay {len(resultado4_df.index)} centros de titularidade pública en Coruña.")
resultado4_df
```

```python
# Seleccionar centros de titularidade pública de Lugo. Cantos hai?

resultado5_df = df[(df['TITULARIDADE'] == "Pública") & (df['Provincia'] == "Lugo")]

print(f"Hay {len(resultado5_df.index)} centros de titularidade pública en Lugo.")
resultado5_df
```

```python
# Seleccionar centros de titularidade pública de Ourense. Cantos hai?

resultado6_df = df[(df['TITULARIDADE'] == "Pública") & (df['Provincia'] == "Ourense")]

print(f"Hay {len(resultado6_df.index)} centros de titularidade pública en Ourense.")
resultado6_df
```

```python
# Seleccionar centros de titularidade pública de Pontevedra. Cantos hai?

resultado7_df = df[(df['TITULARIDADE'] == "Pública") & (df['Provincia'] == "Pontevedra")]

print(f"Hay {len(resultado7_df.index)} centros de titularidade pública en Pontevedra.")
resultado7_df
```

```python
# Seleccionar centros de tipo: CIFP de cada unha das catro provincias. Cantos hai?

resultado8_df = df[(df['Tipo de centro'] == "CIFP")]

print(f"Hay {len(resultado8_df.index)} centros de tipo CIFP.")
resultado8_df
```

```python
# Seleccionar centros de tipo: CPI, CEIP e IES de cada unha das catro provincias. Cantos hai?

resultado9_df = df[(df['Tipo de centro'] == "CPI") | (df['Tipo de centro'] == "CEIP") | (df['Tipo de centro'] == "IES")]

print(f"Hay {len(resultado9_df.index)} centros de titularidade pública en Pontevedra.")
resultado9_df
```

```python
# Seleccionar o número de centros por concello.

centros_por_concello = df.groupby('Concello').size().reset_index(name='Número de centros')

print(centros_por_concello)
```

```python
# Seleccionar o número de centros de cada tipo por concello.

centros_por_tipo_y_concello = df.groupby(['Concello', 'Tipo de centro']).size().reset_index(name='Número de centros')

print(centros_por_tipo_y_concello)
```

```python

# Limpando datos con erros: Dime qué podes limpar e como nestes datos.



df.rename(columns={'Código': 'Codigo', 'Cód. postal': 'Cod.postal','Teléfono':'Telefono','Tipo de centro':'Tipo_de_centro'}, inplace=True)
df = df.dropna(axis="columns", how="all")

# Limpar os datos con valor NA e renomear o header para no ten problemas cos nomes 

```

Garda o dataframe total resultante nos seguintes formatos (e logo volve abrilo nese formato e mostra a cabeceira):

**CSV**

```python
arquivo_temporal = os.path.join(tempfile.mktemp())

centros_csv = df.to_csv(arquivo_temporal,index=False)

print(pd.read_csv(arquivo_temporal).columns.values.tolist())

pd.read_csv(arquivo_temporal).head()
```

**Excel**

Pasa a formato excel.

```python


arquivo_temporal = os.path.join(tempfile.mktemp(suffix=".xlsx"))

centros_excel = df.to_excel(arquivo_temporal,index=False)
pd.read_excel(arquivo_temporal)
```

**JSON**

Pasa a formato JSON.

(Coidado cos índices se xuntaches os dataframes)

```python
df = df.reset_index(drop=True)

arquivo_temporal = os.path.join(tempfile.mktemp(suffix=".json"))

centros_json = df.to_json(arquivo_temporal,index=False)

pd.read_json(arquivo_temporal)



```

**XML**

Pasa a formato XML:

```python
df.reset_index(drop=True, inplace=True)

arquivo_temporal = os.path.join(tempfile.mktemp(suffix=".xml"))

centros_xml = df.to_xml(arquivo_temporal,index=False)

pd.read_xml(arquivo_temporal)

```
