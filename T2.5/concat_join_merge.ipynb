```python
import pandas as pd

# Crear DataFrames de ejemplo
data1 = {
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Juan', 'Luis', 'Marta']
}
df1 = pd.DataFrame(data1)


data2 = {
    'ID': [3, 4, 5, 6],
    'Edad': [23, 34, 45, 56]
}
df2 = pd.DataFrame(data2)

# Uso de merge
merge_inner = pd.merge(df1, df2, on='ID', how='inner')
merge_outer = pd.merge(df1, df2, on='ID', how='outer')
merge_left = pd.merge(df1, df2, on='ID', how='left')
merge_right = pd.merge(df1, df2, on='ID', how='right')

# Mostrar resultados del merge
print('Merge Inner:\n', merge_inner)
print('Merge Outer:\n', merge_outer)
print('Merge Left:\n', merge_left)
print('Merge Right:\n', merge_right)

# Crear DataFrames adicionales para concat

data3 = {
    'ID': [7, 8],
    'Nombre': ['Sofía', 'Carlos']
}
df3 = pd.DataFrame(data3)

# Uso de concat
concat_axis0 = pd.concat([df1, df3], axis=0)
concat_axis1 = pd.concat([df1, df2], axis=1)

# Mostrar resultados del concat
print('Concat axis 0 (filas):\n', concat_axis0)
print('Concat axis 1 (columnas):\n', concat_axis1)

# Uso de join
# Establecer ID como índice
joined = df1.set_index('ID').join(df2.set_index('ID'), how='inner')
print('Join (inner):\n', joined)

# Explicación breve:
# - merge: combina DataFrames basado en una o más claves.
# - concat: concatena DataFrames a lo largo de un eje.
# - join: método simplificado para combinar DataFrames usando el índice.
```
